# Queryes

```python
cur = con.cursor()
select_query1 = "SELECT t.item_key , t.time_key , t.coustomer_key , SUM(t.total_price) " \
    "FROM ecomdb_star_schema.fact_table t " \
    "GROUP BY CUBE(t.item_key, t.time_key, t.coustomer_key) "
cur.execute(select_query1)
records1 = cur.fetchall()
records1

query1 = pd.DataFrame(list(records1), columns=['item_key', 'time_key', 'coustomer_key','SUM'])
query1
```

```python
cur2 = con.cursor()
select_query2 = "SELECT  r.month , SUM(t.total_price)  " \
    "FROM ecomdb_star_schema.fact_table t  " \
       "JOIN ecomdb_star_schema.time_dim r ON t.time_key = r.time_key " \
          "  GROUP BY CUBE(r.month) " \
           "ORDER BY r.month ASC;"
cur2.execute(select_query2)
records2 = cur2.fetchall()
records2

query2 = pd.DataFrame(list(records2), columns=['month','SUM'])
query2
```

```python
cur3 = con.cursor()
select_query3 = "SELECT  r.bank_name , SUM(t.total_price)" \
    "FROM ecomdb_star_schema.fact_table t " \
    	"JOIN ecomdb_star_schema.trans_dim r ON t.payment_key = r.payment_key" \
	       " GROUP BY CUBE(r.bank_name)" \
			  "ORDER BY r.bank_name ASC;"
cur3.execute(select_query3)
records3 = cur3.fetchall()
records3

query3 = pd.DataFrame(list(records2), columns=['Bank_name','SUM'])
query3
```

```python
cur4 = con.cursor()
select_query4 = "SELECT  s.district, t.item_key, SUM(t.total_price) " \
                    "FROM ecomdb_star_schema.fact_table t " \
                  	"JOIN ecomdb_star_schema.store_dim s on s.store_key = t.store_key " \
                    " WHERE  s.district = 'BARISAL' " \
	                "GROUP BY CUBE(s.district, t.item_key) " \
		        	  " ORDER BY s.district ASC;"
cur4.execute(select_query4)
records4 = cur4.fetchall()
records4

query4 = pd.DataFrame(list(records4), columns=['district','item_key','SUM'])
query4
```

```python
cur5 = con.cursor()
select_query5 = "SELECT  s.year, t.item_key, SUM(t.total_price) " \
                    " FROM ecomdb_star_schema.fact_table t " \
                  	" JOIN ecomdb_star_schema.time_dim s on s.time_key = t.time_key " \
	                " WHERE s.year = 2015 " \
		        	" GROUP BY CUBE(s.year, t.item_key) " \
                    " ORDER BY s.year ASC;"
cur5.execute(select_query5)
records5 = cur5.fetchall()
records5

query5 = pd.DataFrame(list(records5), columns=['year','item_key','SUM'])
query5
```

```python
cur6 = con.cursor()
select_query6 = "SELECT  s.year,r.division, t.item_key, SUM(t.total_price) " \
         " FROM ecomdb_star_schema.fact_table t " \
        " JOIN ecomdb_star_schema.time_dim s on s.time_key = t.time_key " \
		" JOIN ecomdb_star_schema.store_dim r on r.store_key = r.store_key " \
		" WHERE (s.year = 2015  AND  r.division = 'BARISAL') " \
        " GROUP BY CUBE(s.year,r.division, t.item_key) " \
        "ORDER BY s.year ASC;" 
cur6.execute(select_query6)
records6 = cur6.fetchall()
records6

query6 = pd.DataFrame(list(records6), columns=['year','division','item_key','SUM'])
query6
```
