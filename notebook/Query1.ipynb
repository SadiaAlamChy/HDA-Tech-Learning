```python
con = PostgresConnection().getConnection()
```

```python
cur = con.cursor()
select_stmt = '''
SELECT s.division, s.district , td.month, td.year, SUM(t.total_price)
FROM star_schema.fact_table t
JOIN star_schema.store_dim s on s.store_key = t.store_key
JOIN star_schema.time_dim td on td.time_key = t.time_key
GROUP BY CUBE(s.division, s.district , td.month, td.year)
'''
cur.execute(select_stmt)
records = cur.fetchall()
records
```

```python
df = pd.DataFrame(list(records), columns=['Division', 'District', 'Month', 'Year','Total Sales'])
df
df = df.dropna()
```
