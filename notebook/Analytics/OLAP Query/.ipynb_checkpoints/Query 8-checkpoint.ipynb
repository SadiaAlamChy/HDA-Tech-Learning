### Q8: What season(quarter) is the worst for each product item?

```python
con = PostgresConnection().getConnection()
cur = con.cursor()
select_stmt ='''WITH grouped_sales AS (SELECT quarter, item_key, SUM(quantity) AS total_quantity
               FROM ecomdb_star_schema.fact_table f
               JOIN ecomdb_star_schema.Time_dim t ON t.time_key = f.time_key
               GROUP BY item_key,quarter) 
               SELECT CONCAT(s.item_key, '->',s.quarter, '(',total_quantity,')') 
               FROM grouped_sales s 
               JOIN (SELECT item_key, min(total_quantity) AS minq
               FROM grouped_sales s GROUP BY item_key) ss ON ss.item_key = s.item_key and s.total_quantity = ss.minq '''
cur.execute(select_stmt)
records = cur.fetchall()
records
```

```python
df = pd.DataFrame(list(records), columns=[ 'Item_ID -> Quarter (Quantity)'])
df
```

```python
con = PostgresConnection().getConnection()
cur = con.cursor()
select_stmt = '''Select  t.item_name,t.quarter,MIN(t.Total)
                    from (Select tim.quarter, i.item_name, SUM(f.quantity) as Total
                    From ecomdb_star_schema.fact_table f
                    JOIN ecomdb_star_schema.item_dim i on i.item_key=f.item_key
                    JOIN ecomdb_star_schema.time_dim tim on tim.time_key=f.time_key
                    GROUP BY (tim.quarter,i.item_name)
                    ORDER BY (tim.quarter)) t
                GROUP BY (t.quarter,t.item_name)
                ORDER BY t.item_name'''
cur.execute(select_stmt)
records = cur.fetchall()
records
```

```python
df['Worst Values'] = pd.to_numeric(df['Worst Values'],errors='coerce')
```

```python
df2 = pd.DataFrame(list(records), columns=['Item_name', 'Quarter' ,'Worst Values'])
df2.head(10)
```

```python
id_vars= ['Item_name']
df_min = df2.groupby(id_vars)['Worst Values'].min()
#df = df2.merge(df_min, how='left',on=id_vars)
df_min
```

```python
con = PostgresConnection().getConnection()
cur = con.cursor()
query = "select i.item_name, td.quarter, sum(f.total_price) "\
"from ecomdb_star_schema.fact_table f "\
"join ecomdb_star_schema.item_dim i on i.item_key=f.item_key "\
"join ecomdb_star_schema.time_dim td on td.time_key = f.time_key "\
"group by i.item_name, td.quarter "\
"order by i.item_name, sum(f.total_price)"
cur.execute(query)
result = cur.fetchall()
result
```

```python
pd_data = pd.DataFrame(list(result), columns=['Item', 'Quarter', 'Sales'])
pd_data = pd_data.dropna()
pd_data = pd_data.groupby('Item').head(1)
pd_data['Worst'] = pd_data['Item'] + ' -> ' + pd_data['Quarter']
pd_data['Worst']
```
